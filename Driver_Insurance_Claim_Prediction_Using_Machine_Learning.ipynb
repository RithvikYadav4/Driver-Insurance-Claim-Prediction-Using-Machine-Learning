RANDOM UNDERSAMPLING + CLUSTER CENTROIDS

Random Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_rus, y_rus = rus.fit_resample(X_train, y_train)

print("After Random Undersampling:", y_rus.value_counts())
```

We will also create a 2-dimensional plot function, plot_2d_space, to see the data distribution:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_2d_space(X, y, label='Classes'):
    colors = ['#1F77B4', '#FF7F0E']
    markers = ['o', 's']

    # Ensure X is a DataFrame for consistent indexing
    if not isinstance(X, pd.DataFrame):
        X = pd.DataFrame(X)

    # Check if there are at least two columns to plot
    if X.shape[1] < 2:
        print("Cannot plot in 2D space, X has less than 2 features.")
        return

    plt.figure(figsize=(8, 6))
    for l, c, m in zip(np.unique(y), colors, markers):
        # Correct indexing for pandas DataFrame: filter rows then select columns by position
        plt.scatter(
            X.loc[y == l].iloc[:, 0], # Select first column (index 0) of filtered data
            X.loc[y == l].iloc[:, 1], # Select second column (index 1) of filtered data
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.xlabel(X.columns[0] if isinstance(X, pd.DataFrame) else 'Feature 1')
    plt.ylabel(X.columns[1] if isinstance(X, pd.DataFrame) else 'Feature 2')
    plt.show()
```

```python
plot_2d_space(X, y, 'Imbalanced dataset (First 2 Features)')
plot_2d_space(X_rus, y_rus ,'Random under-sampling (First 2 Features)')
```

Apply ML Models



Decision Tree

```python
from sklearn.metrics import f1_score, roc_auc_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
import numpy as np

def evaluate_dt(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Metrics
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    cm = confusion_matrix(y_test, y_pred)

    print("\nF1 Score:", f1)
    print("AUROC Score:", auc)
    print("\nConfusion Matrix:\n", cm)

    # Confusion matrix heatmap
    plt.figure(figsize=(5, 4))
    plt.imshow(cm, cmap='Blues')
    plt.title("Confusion Matrix")
    plt.colorbar()

    classes = ["Class 0", "Class 1"]
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], 'd'),
                     ha='center', va='center',
                     color='white' if cm[i, j] > thresh else 'black')

    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.tight_layout()
    plt.show()

    # ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, linewidth=2)
    plt.plot([0,1], [0,1], '--')
    plt.title("ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.grid()
    plt.show()

    return f1, auc, cm
```

Train Decision Tree on Random Undersampled Data FIRST

```python
from sklearn.tree import DecisionTreeClassifier

dt_rus = DecisionTreeClassifier()

# Train on undersampled data
dt_rus.fit(X_rus, y_rus)

```

```python
print("\n===== Decision Tree on Random Undersampling =====")
f1_rus, auc_rus, cm_rus = evaluate_dt(dt_rus, X_test, y_test)
```

```python
print("\n========== Logistic Regression (Random UnderSampling) ==========")
rus_f1, rus_auc, rus_cm = evaluate_dt(lr_rus, X_test, y_test)
```

```python
print("\n========== Logistic Regression (Cluster Centroid) ==========")
cc_f1, cc_auc, cc_cm = evaluate_dt(lr_cc, X_test, y_test)
```
